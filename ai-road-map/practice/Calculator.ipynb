Create a simple calculator with the following features.

Prompts user to select an operation from the list:

1. Add
2. Subtract
3. Multiply
4. Divide

On selecting the option, prompt user to enter two numbers.

Perform operation and show the results in a readable format.










```python
my_calc = { "1" : {
                        "Operation" : "Add",
                        "Opeator" : "+"
               },
              "2" : {
                        "Operation" : "Subtract",
                        "Opeator" : "-"
               },
               "3" : {
                        "Operation" : "Multiply",
                        "Opeator" : "*"
               },
               "4" : {
                        "Operation" : "Divide",
                        "Opeator" : "/"
               },
               "0" : {
                        "Operation" : "Clear",
                        "Opeator" : "+"
               },
               "-1" : {
                        "Operation" : "Exit",
                        "Opeator" : ""
               }
}

ops = ["0", "1","2","3","4","-1"]

print(f"Please select an operation from the list:\n")

selected = 100

while True:
  for k, v in my_calc.items():
    print(f"{k}.{v['Operation']}")

  selected = input("\nEnter:")

  if(selected == -1 or selected == "-1"):
    break;
  elif (selected == 0 or selected == "0"):
    continue;
  elif (selected in ops):
    print(f"\nEnter two numbers:")
    num1 = int(input("Number 1:"))
    num2 = int(input("Number 2:"))
    print(f"{num1} {my_calc[selected]['Opeator']} {num2} = {eval(str(num1) + my_calc[selected]['Opeator'] + str(num2))}")
    print("\nPress any key to continue..")
    input()
    continue;
  else:
    print("Invalid operation, please select a valid operation from the list below.\n")
    continue;




```
